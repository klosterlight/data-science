1. Descargar las primeras 4 páginas
2. Agregar columna de género
3. Agregar columna de entidad

```python
def get_data(params=""):
    keep_reading = True
    people_json = []
    index = 0
    while(keep_reading):
        response = requests.get("https://3de3.mx/declarantes.json?page={}{}".format(index, params))
        if(response.status_code != 200):
            keep_reading = False
            break
        if(response.json() == []):
            keep_reading = False
            break
        people_json += response.json()
    declarantes_todos = pd.DataFrame(people_json)
    return declarantes_todos
data = get_data()
data.count()
# requests.get("https://3de3.mx/declarantes.json?page=20").json()
```
